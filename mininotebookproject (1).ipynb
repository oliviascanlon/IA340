```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# assuming df is your DataFrame
df = pd.read_csv('s3://ia340-2024fallsemester/Electric_Vehicle_Population_Data.csv')

# group by Model and calculate the average Electric Range
grouped = df.groupby('Model')['Electric Range'].mean()

# create bar plot
grouped.plot(kind='bar', figsize=(15, 7))

plt.title('Average Electric Range by Model')
plt.xlabel('Model')
plt.ylabel('Average Electric Range')

plt.show()
```
